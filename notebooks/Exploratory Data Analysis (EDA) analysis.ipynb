```python
# To detect outliers, you can use the Z-score method or the Interquartile Range (IQR) method.
from scipy import stats

# Calculate Z-scores for GHI, DNI, DHI, ModA, ModB, WS, and WSgust
data1['GHI_zscore'] = stats.zscore(data1['GHI'])
data1['DNI_zscore'] = stats.zscore(data1['DNI'])
data1['DHI_zscore'] = stats.zscore(data1['DHI'])
data1['ModA_zscore'] = stats.zscore(data1['ModA'])
data1['ModB_zscore'] = stats.zscore(data1['ModB'])
data1['WS_zscore'] = stats.zscore(data1['WS'])
data1['WSgust_zscore'] = stats.zscore(data1['WSgust'])

# Identify outliers as data points with a Z-score greater than 3 or less than -3
outliers = data1[(data1['GHI_zscore'].abs() > 3) |
              (data1['DNI_zscore'].abs() > 3) |
              (data1['DHI_zscore'].abs() > 3) |
              (data1['ModA_zscore'].abs() > 3) |
              (data1['ModB_zscore'].abs() > 3) |
              (data1['WS_zscore'].abs() > 3) |
              (data1['WSgust_zscore'].abs() > 3)]
print(outliers)

```

```python
# Using IQR:
# Calculate the IQR for the selected columns
Q1 = data1[['GHI', 'DNI', 'DHI', 'ModA', 'ModB', 'WS', 'WSgust']].quantile(0.25)
Q3 = data1[['GHI', 'DNI', 'DHI', 'ModA', 'ModB', 'WS', 'WSgust']].quantile(0.75)
IQR = Q3 - Q1

# Identify outliers
outliers_iqr = data1[((data1[['GHI', 'DNI', 'DHI', 'ModA', 'ModB', 'WS', 'WSgust']] < (Q1 - 1.5 * IQR)) | (data1[['GHI', 'DNI', 'DHI', 'ModA', 'ModB', 'WS', 'WSgust']] > (Q3 + 1.5 * IQR))).any(axis=1)]
print(outliers_iqr)
```

```python
import pandas as pd
from scipy.stats import zscore

# Define file paths for the saved datasets
file_paths = {
    'benin-malanville': 'benin-malanville_data.csv',
    'sierraleone-bumbuna': 'sierraleone-bumbuna_data.csv',
    'togo-dapaong_qc': 'togo-dapaong_qc_data.csv'
}

class DataQualityChecker:
    def __init__(self, file_paths):
        self.data = {country: pd.read_csv(path) for country, path in file_paths.items()}
    
    def calculate_summary_statistics(self):
        for country, df in self.data.items():
            print(f"Summary statistics for {country}:")
            
            # Select only numeric columns
            numeric_cols = df.select_dtypes(include='number')
            
            # Calculate summary statistics
            summary_stats = numeric_cols.describe().T
            summary_stats['median'] = numeric_cols.median()
            summary_stats['std_dev'] = numeric_cols.std()
            
            print(summary_stats)
            print("\n")
    
    def check_data_quality(self):
        for country, df in self.data.items():
            print(f"Data quality check for {country}:")
            missing_values = df.isnull().sum()
            print(f"Missing values:\n{missing_values[missing_values > 0]}")
            print("\n")
            
            critical_columns = ['GHI', 'DNI', 'DHI']
            z_scores = df[critical_columns].apply(zscore)
            outliers = z_scores[(z_scores > 3) | (z_scores < -3)].dropna()
            print(f"Outliers:\n{outliers}")
            print("\n")
            
            incorrect_entries = df[(df['GHI'] < 0) | (df['DNI'] < 0) | (df['DHI'] < 0)]
            print(f"Incorrect entries:\n{incorrect_entries}")
            print("\n")

# Create an instance of DataQualityChecker
checker = DataQualityChecker(file_paths)

# Calculate summary statistics and perform data quality check
checker.calculate_summary_statistics()
checker.check_data_quality()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Define file paths for the saved datasets
file_paths = {
    'benin-malanville': 'benin-malanville_data.csv',
    'sierraleone-bumbuna': 'sierraleone-bumbuna_data.csv',
    'togo-dapaong_qc': 'togo-dapaong_qc_data.csv'
}

class TimeSeriesAnalyzer:
    def __init__(self, file_paths):
        self.dataframes = {country: pd.read_csv(path) for country, path in file_paths.items()}

    def plot_time_series(self):
        for country, df in self.dataframes.items():
            # Convert 'Date' column to datetime format if necessary
            df['Date'] = pd.to_datetime(df['Date'])
            df.set_index('Date', inplace=True)

            # Plot time series
            plt.figure(figsize=(14, 7))
            plt.plot(df.index, df['GHI'], label='GHI', color='orange')
            plt.plot(df.index, df['DNI'], label='DNI', color='blue')
            plt.plot(df.index, df['DHI'], label='DHI', color='green')
            plt.plot(df.index, df['Tamb'], label='Tamb', color='red')
            plt.title(f"Time Series Analysis for {country}")
            plt.xlabel('Date')
            plt.ylabel('Values')
            plt.legend()
            plt.grid(True)
            plt.show()

    def evaluate_cleaning_impact(self):
        for country, df in self.dataframes.items():
            if 'Cleaning' in df.columns:
                cleaned_df = df[df['Cleaning'] == 'Yes']
                df['Date'] = pd.to_datetime(df['Date'])
                cleaned_df['Date'] = pd.to_datetime(cleaned_df['Date'])
                df.set_index('Date', inplace=True)
                cleaned_df.set_index('Date', inplace=True)

                plt.figure(figsize=(14, 7))
                plt.plot(df.index, df['ModA'], label='ModA (Original)', color='blue', alpha=0.5)
                plt.plot(cleaned_df.index, cleaned_df['ModA'], label='ModA (Cleaned)', color='red')
                plt.title(f"Impact of Cleaning on ModA for {country}")
                plt.xlabel('Date')
                plt.ylabel('ModA')
                plt.legend()
                plt.grid(True)
                plt.show()

                plt.figure(figsize=(14, 7))
                plt.plot(df.index, df['ModB'], label='ModB (Original)', color='blue', alpha=0.5)
                plt.plot(cleaned_df.index, cleaned_df['ModB'], label='ModB (Cleaned)', color='red')
                plt.title(f"Impact of Cleaning on ModB for {country}")
                plt.xlabel('Date')
                plt.ylabel('ModB')
                plt.legend()
                plt.grid(True)
                plt.show()

# Create an instance of TimeSeriesAnalyzer
time_series_analyzer = TimeSeriesAnalyzer(file_paths)

# Plot time series for each country
time_series_analyzer.plot_time_series()

# Evaluate the impact of cleaning on sensor readings
time_series_analyzer.evaluate_cleaning_impact()
```
